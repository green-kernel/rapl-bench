```python
# Define the column names
column_names = ['time_sec', 'time_usec', 'diff_time', 'energy_change', 'PCK_no']

# Define list of Computer IDs
id_list = [1070158,1070159,1070163,1070175,1070236,1070242,1070258,1070329,1080229]

# Create an empty pandas data frame with the correct types (using the same names for the columns)
variables = {'time_sec':int(),
             'time_usec':int(),
             'diff_time':int(),
             'energy_change':float(),
             'PCK_no':'',
             'ComputerID':"",
             'CPUload':int()}

df = pd.DataFrame(variables, index=[])

# Load data into the empty data frame
for id in id_list:
    print(id,"sample data/out_"+str(id)+"_no_load.csv")
    # Load the CSV with the specified column names
    df_no_load = pd.read_csv("sample data/out_"+str(id)+"_no_load.csv", sep=" ", names=column_names)
    df_no_load['ComputerID'] = id
    df_no_load['CPUload'] = 0
    df_no_load = df_no_load.drop([0])

    df_stressngc1_load = pd.read_csv("sample data/out_"+str(id)+"_stressngc1_load.csv", sep=" ", names=column_names)
    df_stressngc1_load['ComputerID'] = id
    df_stressngc1_load['CPUload'] = 100
    df_stressngc1_load = df_stressngc1_load.drop([0])
    df = pd.concat([df,df_no_load,df_stressngc1_load], ignore_index=True)
    
    
# Correct the weird counter error for diff_time:
# df[df['diff_time'] < 0]['diff_time'] = df[df['diff_time'] < 0]['diff_time']+1000000

df.loc[df['diff_time'] < 0, 'diff_time'] += 1000000
```

Is the time delay similar for each computer (same model) or is there a significant difference?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot a boxplot for diff_time across different computers
plt.figure(figsize=(12, 6))
sns.boxplot(x='ComputerID', y='diff_time', data=df)

# Add labels and title
plt.title('Distribution of Time Delays (diff_time) Across Computers')
plt.xlabel('ComputerID')
plt.ylabel('Time Delay (diff_time)')

# Display the plot
plt.show()

# Violin plot for diff_time across different computers
plt.figure(figsize=(12, 6))
sns.violinplot(x='ComputerID', y='diff_time', data=df)

# Add labels and title
plt.title('Distribution of Time Delays (diff_time) Across Computers')
plt.xlabel('ComputerID')
plt.ylabel('Time Delay (diff_time)')

# Display the plot
plt.show()
```

There are visual differences, are they different enough to state statistical significance?

```python
from scipy.stats import kruskal

# Perform Kruskal-Wallis test
kruskal_result = kruskal(*[df[df['ComputerID'] == comp_id]['diff_time'] for comp_id in df['ComputerID'].unique()])

# Print the result
print(f'Kruskal-Wallis H-statistic: {kruskal_result.statistic}, p-value: {kruskal_result.pvalue}')
```

**There is a Significant Difference in Delay Times Across Computers:**

The Kruskal-Wallis H-statistic: 374.95is a large test statistic, indicating substantial differences between the groups (i.e., the ComputerIDs).

The p-value: 4.248352677552654e-76 is extremely small, far below the typical significance threshold of 0.05. This means that we can reject the null hypothesis that the distributions of diff_time are the same across all computers.

There seems to be a visual diffeence. Is there a statistical difference?

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, mannwhitneyu, kstest, norm, anderson

# Get the time delay values for CPUload 0 and CPUload 100
no_load = df[df['CPUload'] == 0]['diff_time']
stress_load = df[df['CPUload'] == 100]['diff_time']

# Perform Anderson-Darling test for normality
result_no_load_ad = anderson(no_load, dist='norm')
result_stress_load_ad = anderson(stress_load, dist='norm')

# Print results of the Anderson-Darling test
print("Anderson-Darling Test Results:")
print(f"No load A-D statistic: {result_no_load_ad.statistic}, Critical values: {result_no_load_ad.critical_values}")
print(f"Stress load A-D statistic: {result_stress_load_ad.statistic}, Critical values: {result_stress_load_ad.critical_values}")

# Perform Kolmogorov-Smirnov test for normality
stat_no_load_ks, p_value_no_load_ks = kstest(no_load, 'norm', args=(np.mean(no_load), np.std(no_load)))
stat_stress_load_ks, p_value_stress_load_ks = kstest(stress_load, 'norm', args=(np.mean(stress_load), np.std(stress_load)))

# Print results of the Kolmogorov-Smirnov test
print("\nKolmogorov-Smirnov Test Results:")
print(f"No load K-S statistic: {stat_no_load_ks}, p-value: {p_value_no_load_ks}")
print(f"Stress load K-S statistic: {stat_stress_load_ks}, p-value: {p_value_stress_load_ks}")

# Interpret normality results:
# Anderson-Darling interpretation: if the statistic is lower than the critical value at 5%, we assume normality
if result_no_load_ad.statistic < result_no_load_ad.critical_values[2] and result_stress_load_ad.statistic < result_stress_load_ad.critical_values[2]:
    print("\nBoth groups are normally distributed according to the Anderson-Darling test.")
else:
    print("\nAt least one group is not normally distributed according to the Anderson-Darling test.")

# Kolmogorov-Smirnov interpretation: p-value > 0.05 means data is normally distributed
if p_value_no_load_ks > 0.05 and p_value_stress_load_ks > 0.05:
    print("Both groups are normally distributed according to the Kolmogorov-Smirnov test.")
else:
    print("At least one group is not normally distributed according to the Kolmogorov-Smirnov test.")

# If both groups pass normality tests, perform t-test, otherwise Mann-Whitney U test
if (result_no_load_ad.statistic < result_no_load_ad.critical_values[2] and 
    result_stress_load_ad.statistic < result_stress_load_ad.critical_values[2] and 
    p_value_no_load_ks > 0.05 and p_value_stress_load_ks > 0.05):
    # Perform independent t-test
    t_stat, p_value = ttest_ind(no_load, stress_load, equal_var=False)
    print(f'\nIndependent t-test p-value: {p_value}')
    if p_value < 0.05:
        print("There is a significant difference between the time delays.")
    else:
        print("There is no significant difference between the time delays.")
else:
    # Perform Mann-Whitney U test
    stat, p_value = mannwhitneyu(no_load, stress_load)
    print(f'\nMann-Whitney U test p-value: {p_value}')
    if p_value < 0.05:
        print("There is a significant difference between the time delays.")
    else:
        print("There is no significant difference between the time delays.")
```

**There is a significant difference between the time delays between CPU load ~0 and 100**

The statistical tests:

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, mannwhitneyu, kstest, norm, anderson

# Get the unique ComputerIDs
computer_ids = df['ComputerID'].unique()

# Loop through each ComputerID and perform the normality tests and appropriate comparison tests
for comp_id in computer_ids:
    print(f"\n--- Testing for ComputerID: {comp_id} ---")
    
    # Get data for CPUload 0 and 100 for the current ComputerID
    no_load = df[(df['ComputerID'] == comp_id) & (df['CPUload'] == 0)]['diff_time']
    stress_load = df[(df['ComputerID'] == comp_id) & (df['CPUload'] == 100)]['diff_time']
    
    # Check if there are enough samples for the tests
    if len(no_load) < 3 or len(stress_load) < 3:
        print(f"Not enough data for ComputerID {comp_id}, skipping tests.")
        continue
    
    # Perform Anderson-Darling test for normality
    result_no_load_ad = anderson(no_load, dist='norm')
    result_stress_load_ad = anderson(stress_load, dist='norm')

    # Perform Kolmogorov-Smirnov test for normality
    stat_no_load_ks, p_value_no_load_ks = kstest(no_load, 'norm', args=(np.mean(no_load), np.std(no_load)))
    stat_stress_load_ks, p_value_stress_load_ks = kstest(stress_load, 'norm', args=(np.mean(stress_load), np.std(stress_load)))
    
    # Print results of the Anderson-Darling test
    print(f"Anderson-Darling Test Results for ComputerID {comp_id}:")
    print(f"  No load A-D statistic: {result_no_load_ad.statistic}, Critical values: {result_no_load_ad.critical_values}")
    print(f"  Stress load A-D statistic: {result_stress_load_ad.statistic}, Critical values: {result_stress_load_ad.critical_values}")

    # Print results of the Kolmogorov-Smirnov test
    print(f"Kolmogorov-Smirnov Test Results for ComputerID {comp_id}:")
    print(f"  No load K-S statistic: {stat_no_load_ks}, p-value: {p_value_no_load_ks}")
    print(f"  Stress load K-S statistic: {stat_stress_load_ks}, p-value: {p_value_stress_load_ks}")

    # Check normality results and perform appropriate test
    normal_ad = (result_no_load_ad.statistic < result_no_load_ad.critical_values[2] and 
                 result_stress_load_ad.statistic < result_stress_load_ad.critical_values[2])
    normal_ks = (p_value_no_load_ks > 0.05 and p_value_stress_load_ks > 0.05)
    
    if normal_ad and normal_ks:
        # Perform independent t-test if both groups pass the normality tests
        t_stat, p_value = ttest_ind(no_load, stress_load, equal_var=False)
        print(f'\nIndependent t-test p-value for ComputerID {comp_id}: {p_value}')
        if p_value < 0.05:
            print(f"There is a significant difference in time delays for ComputerID {comp_id}.")
        else:
            print(f"No significant difference in time delays for ComputerID {comp_id}.")
    else:
        # Perform Mann-Whitney U test if either group fails the normality tests
        stat, p_value = mannwhitneyu(no_load, stress_load)
        print(f'\nMann-Whitney U test p-value for ComputerID {comp_id}: {p_value}')
        if p_value < 0.05:
            print(f"There is a significant difference in time delays for ComputerID {comp_id}.")
        else:
            print(f"No significant difference in time delays for ComputerID {comp_id}.")
```

**There is a general pattern that across most computers, there is a significant difference in time delays (diff_time) when the CPUload changes from 0 to 100. This suggests that CPU load impacts time delays, but the extent of this impact varies between machines.**

There is a Significant differences in diff_time between CPUload 0 and 100 for most ComputerIDs. These include:
1070158, 1070159, 1070163, 1070236, 1070242, 1070258, 1070329, 1080229: The p-values for these computers are all below 0.05, indicating that there is a significant difference in time delays when the CPUload changes.

However, for ComputerID 1070175, the p-value is 0.15, which is above the standard threshold of 0.05. This indicates that there is no significant difference in time delays for this particular computer when switching between CPUload 0 and 100.

```python
from scipy.stats import kruskal

# Loop through each unique combination of CPUload and ComputerID
grouped_data = []

for cpu_load in df['CPUload'].unique():
    for comp_id in df['ComputerID'].unique():
        # Append diff_time for each combination of CPUload and ComputerID
        grouped_data.append(df[(df['CPUload'] == cpu_load) & (df['ComputerID'] == comp_id)]['diff_time'])

# Perform Kruskal-Wallis test
kruskal_result = kruskal(*grouped_data)

# Print the result
print(f'Combined CPUload and ComputerID: Kruskal-Wallis H-statistic: {kruskal_result.statistic}, p-value: {kruskal_result.pvalue}')
```
